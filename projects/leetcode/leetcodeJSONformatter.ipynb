```python
def addDigits(num):
    """
    :type num: int
    :rtype: int
    """
    while(num > 9):
        total = 0
        num = str(num)
        for n in num:
            total += int(n)
        num = total
    return num
for i in range(100):
    print('{} {}'.format(i,addDigits(i)))
```
